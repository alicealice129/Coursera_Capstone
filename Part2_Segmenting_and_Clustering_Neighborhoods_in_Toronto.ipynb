## Part1

- Obtain the data that is in the table of postal codes and transform the data into a pandas dataframe. 

    - The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood
    - Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
    - More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma.
    - If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
    - Clean Notebook and add Markdown cells to explain the work and any assumptions.
    - In the last cell of notebook, use the .shape method to print the number of rows of your dataframe.

```python
# import library
import pandas as pd
```

```python
toronto_wiki_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
dfs = pd.read_html(toronto_wiki_url, header=0)  # set header with row #0
```

```python
# show dataframe
dfs[0].head(12)
```

- Define the dataframe columns

```python
df = dfs[0].rename(columns ={'Postal code': 'PostalCode'}) 
df.head(12)
```

```python
df.shape
```

- Ignore cells with a borough that is 'Not assigned'.

```python
rows_to_drop = df.index[df['Borough'] == 'Not assigned']
rows_to_drop
```

```python
df.drop(rows_to_drop, inplace=True)
df.head(12)
```

```python
# reset index 
df_toronto = df.reset_index(
    drop=True,  # do not save original index as column
)
df_toronto.head(12)
```

- Two or more rows will be combined into one row with the neighborhoods separated with a comma

```python
df_toronto['Neighborhood'] = df_toronto['Neighborhood'].str.replace(' /', ',')
df_toronto.head(12)
```

```python
df_toronto.shape
```
